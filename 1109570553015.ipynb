```python

from numpy.random import seed
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split
from keras.layers import Input, Dense
from keras.models import Model
```

```python
sX = minmax_scale(X, axis = 0)
ncol = sX.shape[1]
X_train, X_tes, Y_train, Y_tes = train_test_split(sX, Y, train_size = 1, random_state = seed(2017))
```

```python
input_dim = Input(shape = (ncol, ))
encoding_dim = 15
encoded1 = Dense(110, activation = 'relu')(input_dim)
encoded2 = Dense(95, activation = 'relu')(encoded1)
encoded3 = Dense(70, activation = 'relu')(encoded2)
encoded4 = Dense(55, activation = 'relu')(encoded3)
encoded5 = Dense(30, activation = 'relu')(encoded4)
encoded6 = Dense(encoding_dim, activation = 'relu')(encoded5)
decoded1 = Dense(15, activation = 'relu')(encoded4)
decoded2 = Dense(30, activation = 'relu')(decoded1)
decoded3 = Dense(55, activation = 'relu')(decoded2)
decoded4 = Dense(70, activation = 'relu')(decoded3)
decoded5 = Dense(95, activation = 'relu')(encoded4)
decoded6 = Dense(ncol, activation = 'sigmoid')(decoded5)
autoencoder = Model(input = input_dim, output = decoded6)
autoencoder.compile(optimizer = 'adadelta', loss = 'binary_crossentropy')
autoencoder.fit(X_train, X_train, nb_epoch = 100, batch_size = 100, shuffle = True, validation_data = (X_tes, X_tes))
encoder = Model(input = input_dim, output = encoded6)
encoded_input = Input(shape = (encoding_dim, ))
encoded_out = encoder.predict(X_tes)

```

```python
from sklearn.neighbors import KNeighborsClassifier 
clf = KNeighborsClassifier(n_neighbors = 3)  
clf.fit(encoded_out, Y_tes)

```
